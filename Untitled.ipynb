```python
df['Year'] = df['Date Reported'].dt.year
```

```python
crime_counts_by_year = df['Year'].value_counts().sort_index()
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(crime_counts_by_year.index, crime_counts_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.title('Crime Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
crime_counts = df.groupby(['Year', 'Index Crime Category']).size().unstack(fill_value=0)
```

```python
# Transpose the DataFrame for plotting
crime_counts = crime_counts.T

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Define the width of each bar group
bar_width = 0.2

# Create an array of x values for each group of bars
x = np.arange(len(crime_counts.columns))

# Loop through each year and plot a bar group
for i, year in enumerate(crime_counts.index):
    ax.bar(x + i * bar_width, crime_counts.loc[year], width=bar_width, label=year)

# Set the x-axis labels and title
ax.set_xticks(x + (len(crime_counts.index) / 2 - 0.5) * bar_width)
ax.set_xticklabels(crime_counts.columns)
ax.set_xlabel('Crime Category')
ax.set_ylabel('Number of Crimes')
ax.set_title('Crime Categories Over the Years')

# Add a legend
ax.legend(title='Year', loc='upper right')

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
crime_counts = df['Statute or Ordinance Description'].value_counts()
```

```python
# Data for the pie chart
labels = top_5_crimes.index
sizes = top_5_crimes.values

# Explode the first slice (i.e., 'explode' the largest slice)
explode = (0.1, 0, 0, 0, 0)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Top 5 Crimes by Description')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Get the top 5 crimes by count
top_5_crimes = df['Statute or Ordinance Description'].value_counts().head(5).index.tolist()

# Filter the DataFrame to include only the top 5 crimes
filtered_df = df[df['Statute or Ordinance Description'].isin(top_5_crimes)]

# Group the filtered data by 'Year' and 'Statute or Ordinance Description' and count occurrences
crime_counts = filtered_df.groupby(['Year', 'Statute or Ordinance Description']).size().unstack(fill_value=0)

# Create the comparative grouped bar chart
fig, ax = plt.subplots(figsize=(12, 8))

# Define the width of each bar
bar_width = 0.15

# Create an array of x values for each group of bars
x = np.arange(len(crime_counts.index))

# Loop through the top 5 crimes and plot a grouped bar for each crime by year
for i, crime in enumerate(top_5_crimes):
    ax.bar(x + i * bar_width, crime_counts[crime], width=bar_width, label=crime)

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Number of Crimes')
ax.set_title('Top 5 Crimes by Year (Grouped Bar Chart)')

# Add a legend
ax.legend()

# Set the x-axis ticks and labels
ax.set_xticks(x + (len(top_5_crimes) / 2 - 0.5) * bar_width)
ax.set_xticklabels(crime_counts.index)

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Population data
population_data = {
    'Year': [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Population': [457715, 463555, 475936, 483518, 489683, 498879, 489109, 495298]
}

# Create a DataFrame for population
population_df = pd.DataFrame(population_data)

# Merge population data with the patrol_df DataFrame
merged_patrol_df = pd.merge(patrol_df, population_df, on='Year')
```

```python
# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(12, 8))

# Create a bar chart for crime counts by year
ax1.plot(crime_counts_by_year.index, crime_counts_by_year.values, label='Crime Counts', color='b', alpha=0.7)

# Set labels and title for the bar chart
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Crimes')
ax1.set_title('Crime Counts, Patrol Officers, and Population Over Years')

# Create a secondary axis for Patrol Officers as a line plot
ax2 = ax1.twinx()
ax2.plot(merged_patrol_df['Year'], merged_patrol_df['Patrol Officers'], color='r', marker='o', label='Patrol Officers')

# Set labels for the line plot
ax2.set_ylabel('Number of Patrol Officers', color='r')

# Create a third axis for Population as a line plot
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 80))  # Adjust the position of the third axis
ax3.plot(merged_patrol_df['Year'], merged_patrol_df['Population'], color='g', marker='s', label='Population')

# Set labels for the population line plot
ax3.set_ylabel('Population', color='g')

# Add legends for all three plots
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(lines1 + lines2 + lines3, labels1 + labels2 + labels3, loc='upper left')

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
